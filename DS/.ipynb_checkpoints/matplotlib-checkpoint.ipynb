# Matplotlib
Matplotlib is a low level graph plotting library in python that serves as a visualization utility. Its numerical mathematics extension is NumPy. Most of the Matplotlib utilities lies under the pyplot submodule, and are usually imported under the plt alias.

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Plotting Lines

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

# Plotting the lines connecting the points (xi, yi)
plt.plot(x, y, color='g', linestyle='--', linewidth=3, marker='x', markersize=15, label='My Line')

# Title and labels
plt.title("Plotting Lines")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

# Legend is used to identify the elements in the plot
plt.legend()

plt.show()
```

```python
x = np.arange(0, 10, 1)
y = 3 * x + 5

plt.plot(x, y, color='red', marker='o', markersize='5', label = 'graph')
plt.legend()
plt.show()
```

#### 1. Sine and Cosine

```python
x = np.linspace(-np.pi, np.pi, 100)          # takes 100 points between -pi and pi
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, color='b', label='Sine')     # works even if the color is not mentioned 
plt.plot(x, y2, color='g', label='Cosine')

plt.legend()
plt.show()
```

#### 2. Subplots
Allows us to create multiple plots in a single figure

```python
# Create a figure
plt.figure()

# Creates subplot (rows, cols, current_plot)
plt.subplot(2, 1, 1)
plt.plot(x, y1)
plt.title("Sine Wave")

plt.subplot(2, 1, 2)
plt.plot(x, y2)
plt.title("Cosine Wave")

plt.tight_layout()
plt.show()

```

#### 3. Bar Chart

```python
x = np.arange(1, 6)
y = np.random.randint(1, 50, 5)

plt.bar(x, y, align='center')     # align='edge', center is default
plt.show()
```

```python
plt.title("Red Bars")
plt.bar(np.arange(1, 6), np.random.randint(1, 10, 5), color='red')
plt.show()
```

#### 4. Pie Chart

```python
x = np.array([4, 4, 1])

plt.pie(x, colors=['r','b','orange'], labels=['A','B','C'], explode=[0, 0, 0.1], startangle=10)  # can use () instead of []
plt.show()
```

#### 5. Histogram

```python
data = np.random.normal(0, 1, 100)      # Normal distribution with mean 0 and standard deviation 1, 100  points

plt.hist(data, color='lime', edgecolor='yellow')
plt.show()
```

#### 6. Scatter Plot

```python
x = [5, 7, 8, 7, 2, 17, 2, 9, 4, 11, 12, 9, 6]
y = [99, 86, 87, 88, 100, 86, 103, 87, 94, 78, 77, 85, 86]
```

```python
plt.title('Scatter Plot')
plt.xlabel('X')
plt.ylabel('Y')

plt.scatter(x, y, color='tan', marker='v')
plt.show()
```
