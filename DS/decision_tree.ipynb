```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
dataset = load_iris()
```

```python
X, y = dataset.data, dataset.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
classifier = DecisionTreeClassifier()
```

```python
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)

for i in range(10):
    print(f"Predicted: {dataset.target_names[y_pred[i]]}, Actual: {dataset.target_names[y_test[i]]}")
```

```python
accuracy = accuracy_score(y_pred,y_test)
print(f"Accuracy: {accuracy}")
```

```python
result = classifier.predict([[4, 3, 1, 0]])
print(dataset.target_names[result])
```

```python
tree.plot_tree(classifier)
```

```python
import matplotlib.pyplot as plt

print(y_pred, y_test)
plt.scatter(y_pred, y_test, alpha=0.7, color='b')
plt.title("Actual v/s Predicted Values")
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.grid(True)
plt.axis('equal')
plt.show()
```
