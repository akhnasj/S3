# Matplotlib
Matplotlib is a low level graph plotting library in python that serves as a visualization utility. Its numerical mathematics extension is NumPy. Most of the Matplotlib utilities lies under the pyplot submodule, and are usually imported under the plt alias.

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Plotting Lines

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

# Plotting the lines connecting the points (xi, yi)
plt.plot(x, y, color='g', linestyle='--', linewidth=3, marker='x', markersize=15, label='My Line')

# Title and labels
plt.title("Plotting Lines")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

# Legend is used to identify the elements in the plot
plt.legend()

plt.show()
```

```python
x = np.arange(0, 10, 1)
y = 3 * x + 5

plt.plot(x, y, color='red', marker='o', markersize='5', label = 'graph')
plt.legend()
plt.show()
```
